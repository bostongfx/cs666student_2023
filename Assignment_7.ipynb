**Task 4**: Let's apply a 3x3 filter mask! [30 Points]

```python
# Here we will modify the pixel data!
#
# The following code allows you to set the pixels of the current slice.
#
#   pixels.set(new_pixels);
#
# Write code to APPLY A FILTER MASK to the pixels.
#
# Hint: You can use the following code to get the dimensions of the current slice.
#   w = element.image.width;
#   h = element.image.height;
#
# You can decide which filter mask to apply (e.g., Sobel, Gaussian blur, edge detection..).
#
# After setting new pixel values, you can re-draw/re-render the current slice with this command:
#
#   cornerstone.renderGrayscaleImage(element, true);
#
```

```python
#
#
# TODO: YOUR CODE GOES HERE
#
#

let filter = [
  [1 / 9, 1 / 9, 1 / 9],
  [1 / 9, 1 / 9, 1 / 9],
  [1 / 9, 1 / 9, 1 / 9]
];
let w = element.image.width;
let h = element.image.height;

let new_pixels = new Int16Array(w * h);

for(let y = 1; y < h - 1; y++) {
  for(let x = 1; x < w - 1; x++) {

    let sum = 0;
    for(let fy = -1; fy <= 1; fy++) {
      for(let fx = -1; fx <= 1; fx++) {
        let imageX = x + fx;
        let imageY = y + fy;
        let pixelValue = pixels[imageY * w + imageX];
        let filterValue = filter[fy + 1][fx + 1];
        sum += pixelValue * filterValue;
      }
    }

    new_pixels[y * w + x] = sum;
  }
}

pixels.set(new_pixels);
cornerstone.renderGrayscaleImage(element, true);
```

The main difference between processing with a bookmarklet and kernel mask filtering from above lies in how they access and manipulate pixels.

A bookmarklet is a small piece of JavaScript code that can be saved as a bookmark in a web browser. When clicked, the bookmarklet code is executed on the current web page, allowing you to modify the page's content or interact with it in some way. In the context of image processing, a bookmarklet can access the pixels of an image on a web page and perform various operations on them, such as applying filters or transformations.

On the other hand, kernel mask filtering refers to a specific technique used in image processing algorithms, particularly in convolution operations. In kernel mask filtering, a small matrix called a kernel or mask is applied to each pixel in an image, and the values of the surrounding pixels are used to calculate the new value of the central pixel. This process is often used for operations such as blurring, sharpening, or edge detection.

The key distinction is that bookmarklets operate within the context of a web page and can access the pixels of an image displayed on that page. They typically use JavaScript to access the Document Object Model (DOM) of the page and manipulate the image element directly.

Kernel mask filtering, on the other hand, is a general technique used in image processing algorithms and is not limited to web pages or specific programming languages. It involves applying a kernel to each pixel of an image, using a mathematical operation to calculate the new pixel value based on the surrounding pixels. This technique can be implemented in various programming languages, libraries, and frameworks.

In summary, the main difference is that bookmarklets are specific to web pages and use JavaScript to access and manipulate image pixels directly, while kernel mask filtering is a more general technique used in image processing algorithms to calculate new pixel values based on surrounding pixels.
