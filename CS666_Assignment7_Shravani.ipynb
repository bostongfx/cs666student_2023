**Task 4**: Let's apply a 3x3 filter mask! [30 Points]

```python
# Here we will modify the pixel data!
#
# The following code allows you to set the pixels of the current slice.
#
#   pixels.set(new_pixels);
#
# Write code to APPLY A FILTER MASK to the pixels.
#
# Hint: You can use the following code to get the dimensions of the current slice.
#   w = element.image.width;
#   h = element.image.height;
#
# You can decide which filter mask to apply (e.g., Sobel, Gaussian blur, edge detection..).
#
# After setting new pixel values, you can re-draw/re-render the current slice with this command:
#
#   cornerstone.renderGrayscaleImage(element, true);
#
```

```python

#
#
# TODO: YOUR CODE GOES HERE
#
#

const w = element.image.width;
const h = element.image.height;
const new_pixels = new Float32Array(pixels.length); 

// The Sobel filter masks for horizontal and vertical gradients
const sobelHorizontalMask = [
  -1, 0, 1,
  -2, 0, 2,
  -1, 0, 1
];

const sobelVerticalMask = [
  -1, -2, -1,
  0, 0, 0,
  1, 2, 1
];

// Apply the Sobel filter
for (let y = 1; y < h - 1; y++) {
  for (let x = 1; x < w - 1; x++) {
    let horizontalGradient = 0;
    let verticalGradient = 0;

    let count = 0;

    for (let filterY = -1; filterY <= 1; filterY++) {
      for (let filterX = -1; filterX <= 1; filterX++) {
          
        const currentX = x + filterX;
        const currentY = y + filterY;

        const pixelIndex = currentY * w + currentX;

        horizontalGradient += pixels[pixelIndex] * sobelHorizontalMask[count];
        verticalGradient += pixels[pixelIndex] * sobelVerticalMask[count];

        count++;
      }
    }

    const gradientMagnitude = Math.sqrt(horizontalGradient ** 2 + verticalGradient ** 2);
    new_pixels[y * w + x] = gradientMagnitude;
  }
}

// Set the new pixel values
pixels.set(new_pixels);

// Re-render the current slice
cornerstone.renderGrayscaleImage(element, true);
```
